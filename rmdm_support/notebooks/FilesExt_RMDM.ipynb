```python
import pandas as pd
from cognite.client import CogniteClient
from pathlib import Path
from cognite.client.data_classes.data_modeling import NodeId
```

```python
# make the client object
# this can be done in many different ways, pick the one you feel more comfortable with...

client = CogniteClient()
```

```python
# ---------- CONFIG ----------
ROOT_FOLDER = Path("Files")  # walk recursively
SPACE = "dmu_rmdm_instances"

def file_ext_id(p: Path) -> str:
    # use plain stem or add a prefix like f"FILE-{p.stem}"
    return p.stem

uploaded = 0
for path in ROOT_FOLDER.rglob("*"):
    if not path.is_file():
        continue
    ext_id = file_ext_id(path)
    try:
        client.files.upload_content(
            path=str(path),
            instance_id=NodeId(SPACE, ext_id),
        )
        uploaded += 1
        print(f"✓ Uploaded {path.name} -> {SPACE}:{ext_id}")
    except Exception as e:
        print(f"✗ {path.name} -> {SPACE}:{ext_id} : {e}")

    print(f"Done. Uploaded {uploaded} files.")
```
